```python
x = dataset.iloc[:, 0:21]  # Selecting all rows, and columns from index 1 (inclusive) to the second-to-last column
y = dataset.iloc[:, 21:22]    # Selecting all rows, and the last column
```

# Split into Traning and Testing Data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# NN

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(units=21, activation='tanh')) #input layer
```

### Hidden Layer

```python
# Add the first hidden layer
model.add(Dense(units=10, activation='tanh'))

# Add the second hidden layer
model.add(Dense(units=6, activation='tanh'))

# Add the third hidden layer
model.add(Dense(units=4, activation='tanh'))

model.add(Dense(units=2, activation='softmax'))
```

### Output Layer

```python
# Add the output layer
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss = "binary_crossentropy", optimizer="adam" , metrics= ["accuracy"]) #loss function , accuracy and optimizer 
```

```python
# Train the model
history = model.fit(x_train, y_train, epochs=100, batch_size=64, validation_data=(x_test, y_test))
```

```python
model.save("Phising.h5")
```

```python
from sklearn.metrics import confusion_matrix

# Make predictions on test data
y_pred = model.predict(x_test)

# Convert probabilities to class labels
y_pred_classes = np.argmax(y_pred, axis=1)

# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_classes)

print("Confusion Matrix:")
print(conf_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_classes)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import numpy as np

np.random.seed(42)
x_noisy = np.random.normal(loc=x, scale=0.5) 
y_noisy = y.copy()  # Assuming y is not affected by noise

x_train, x_test, y_train, y_test = train_test_split(x_noisy, y_noisy, test_size=0.2, random_state=42)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.25, random_state=42)  # 0.25 x 0.8 = 0.2

# Create RandomForestClassifier model
rf_model = RandomForestClassifier(random_state=42)

# Train the model
rf_model.fit(x_train, y_train)

# Make predictions on test data
y_pred_rf_test = rf_model.predict(x_test)

# Calculate test accuracy
accuracy_rf_test = accuracy_score(y_test, y_pred_rf_test)
print("Random Forest Classifier Test Accuracy:", accuracy_rf_test)

# Make predictions on validation data
y_pred_rf_val = rf_model.predict(x_val)

# Calculate validation accuracy
accuracy_rf_val = accuracy_score(y_val, y_pred_rf_val)
print("Random Forest Classifier Validation Accuracy:", accuracy_rf_val)

# Create SVM classifier model
svm_model = SVC(kernel='rbf', random_state=42)

# Train the model
svm_model.fit(x_train, y_train)

# Make predictions on test data
y_pred_svm_test = svm_model.predict(x_test)

# Calculate test accuracy
accuracy_svm_test = accuracy_score(y_test, y_pred_svm_test)
print("SVM Classifier Test Accuracy:", accuracy_svm_test)

# Make predictions on validation data
y_pred_svm_val = svm_model.predict(x_val)

# Calculate validation accuracy
accuracy_svm_val = accuracy_score(y_val, y_pred_svm_val)
print("SVM Classifier Validation Accuracy:", accuracy_svm_val)
```
